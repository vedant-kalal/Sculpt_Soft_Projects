# Skewness: Measuring Distribution Asymmetry

---

## 1. The Core Idea
Skewness measures how much a distribution is asymmetric (tilted) instead of being perfectly balanced like a normal bell curve.

- **Symmetric distribution:** mean = median = mode (perfect bell curve).
- **Positive skew (right-skewed):** tail goes to the right (e.g., income data: many low incomes, a few very high ones).
- **Negative skew (left-skewed):** tail goes to the left (e.g., test scores when most people score high but a few very low).

---

## 2. The Math (but easy)
The formula for skewness is:

$$
\text{Skewness} = \frac{1}{n} \sum_{i=1}^n \frac{(x_i - \mu)^3}{\sigma^3}
$$

Where:
- $x_i$ = data points
- $\mu$ = mean
- $\sigma$ = standard deviation
- $n$ = number of data points

The cube $(x_i - \mu)^3$ is the trick:
- Positive values → long right tail (positive skew).
- Negative values → long left tail (negative skew).

---

## 3. Intuition with Examples
- Exam scores where most students fail, but a few score very high → right skew (tail to the right).
- Exam scores where most students get nearly perfect scores, but a few bomb → left skew (tail to the left).

---

## 4. Why it Matters in ML
Many algorithms (e.g., Linear Regression, Logistic Regression) assume input features are normally distributed (symmetric).

Strong skewness can hurt performance.

**Common fixes:**
- Apply log transform, square root, or Box-Cox transform to reduce skew.
- Standardize features to make them more Gaussian-like.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate three datasets: normal, positively skewed, negatively skewed
np.random.seed(42)

# Normal distribution
data_normal = np.random.normal(0, 1, 1000)

# Positively skewed (exponential)
data_pos_skew = np.random.exponential(1, 1000)

# Negatively skewed (by flipping exponential)
data_neg_skew = -np.random.exponential(1, 1000)

# Calculate skewness values
skew_normal = skew(data_normal)
skew_pos = skew(data_pos_skew)
skew_neg = skew(data_neg_skew)

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Normal
axes[0].hist(data_normal, bins=30, color="skyblue", edgecolor="black")
axes[0].set_title(f"Normal (Skew={skew_normal:.2f})")

# Positive skew
axes[1].hist(data_pos_skew, bins=30, color="lightgreen", edgecolor="black")
axes[1].set_title(f"Positive Skew (Skew={skew_pos:.2f})")

# Negative skew
axes[2].hist(data_neg_skew, bins=30, color="salmon", edgecolor="black")
axes[2].set_title(f"Negative Skew (Skew={skew_neg:.2f})")

plt.show()
```
