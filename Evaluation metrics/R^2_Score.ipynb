# R² Score: Measuring Regression Model Fit

---

## 1. Where R² is Used
Unlike accuracy, precision, recall, or ROC (which are for classification), **R²** (coefficient of determination) is for regression problems (predicting numbers, like house prices, temperatures, exam scores, etc.).

It tells us how well our regression model explains the variation in the data.

---

## 2. Formula
The R² score is:

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

Where:
- $SS_{res} = \sum (y_i - \hat{y}_i)^2$  → Residual sum of squares (errors left after the model).
- $SS_{tot} = \sum (y_i - \bar{y})^2$  → Total sum of squares (variation if you just predicted the mean every time).
- $y_i$ = actual values
- $\hat{y}_i$ = predicted values
- $\bar{y}$ = mean of actual values

---

## 3. Intuition
- If model predicts perfectly, then $SS_{res} = 0$, so $R^2 = 1$.
- If model is as bad as just guessing the mean, then $SS_{res} = SS_{tot}$, so $R^2 = 0$.
- If model is worse than mean, then $R^2 < 0$.

---

## 4. Small Example
Suppose actual $y = [3, 5, 7]$, predicted $\hat{y} = [2.8, 4.9, 7.1]$.

Compute mean:
$\bar{y} = (3 + 5 + 7)/3 = 5$

$SS_{tot} = (3-5)^2 + (5-5)^2 + (7-5)^2 = 4 + 0 + 4 = 8$

$SS_{res} = (3-2.8)^2 + (5-4.9)^2 + (7-7.1)^2 = 0.04 + 0.01 + 0.01 = 0.06$

$R^2 = 1 - \frac{0.06}{8} = 1 - 0.0075 = 0.9925$

→ Very close to 1, so model fits well.


```python
# Example: Calculating R² Score in Python

import numpy as np
from sklearn.metrics import r2_score

y_true = np.array([3, 5, 7])
y_pred = np.array([2.8, 4.9, 7.1])

# Manual calculation
mean_y = np.mean(y_true)
ss_tot = np.sum((y_true - mean_y) ** 2)
ss_res = np.sum((y_true - y_pred) ** 2)
r2_manual = 1 - (ss_res / ss_tot)
print(f"R² (manual): {r2_manual:.4f}")

# Using sklearn
r2_sklearn = r2_score(y_true, y_pred)
print(f"R² (sklearn): {r2_sklearn:.4f}")
```

```python
# Visualization: R² Score with Actual vs Predicted Plot
import matplotlib.pyplot as plt
import numpy as np

# Use the same y_true and y_pred as above
# y_true = np.array([3, 5, 7])
# y_pred = np.array([2.8, 4.9, 7.1])

plt.figure(figsize=(6, 5))
plt.scatter(y_true, y_pred, color='blue', label='Predicted vs Actual')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('R² Score Visualization')
plt.legend()
plt.grid(True)
plt.show()
```
