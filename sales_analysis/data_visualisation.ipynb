```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv("cleaned_sales_data.csv")
df.head()
```

```python
sb.set_style("whitegrid")
```

```python
sb.displot(df['Revenue'], kde=True, color='blue', bins=30)
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')   
plt.ylabel('Frequency')
plt.show()
```

```python
sb.displot(df["Is_Weekend"], color='blue', bins=30)
plt.title('weekend distribution')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()
```

```python
sb.jointplot(x='Revenue', y=df["Units_Sold"], data=df, color='blue')
plt.title('Joint Plot of Revenue and Weekend_days')
plt.xlabel('Revenue')   
plt.ylabel('Is_Weekend')
plt.show()
```

```python
sb.lmplot(x='Revenue', y='Units_Sold', data=df,palette='coolwarm', height=6, aspect=1.5)
plt.title('Regression Plot of Revenue and Units Sold')
plt.xlabel('Revenue')
plt.ylabel('Units Sold')
plt.show()
```

```python
sb.jointplot(x='Revenue', y=df["Avg_Selling_Price"], data=df, color='blue')
plt.xlabel('Revenue')   
plt.ylabel('Avg_Selling_Price')
plt.show()
```

```python
sb.pairplot(df, hue='Is_Weekend', palette='coolwarm')
plt.show()
```

```python
figure= plt.figure(figsize=(10, 6))
sb.barplot(x=df['Month'],y= df['Revenue'], color='blue')
plt.title('Monthly Revenue')
plt.xlabel('Month') 
plt.ylabel('Revenue')
plt.show()
```

```python
figure= plt.figure(figsize=(10, 6))
sb.countplot(x='Month', data=df, palette='coolwarm',hue='Is_Weekend')
plt.title("count weekend days by month")
plt.xlabel('Revenue')
plt.ylabel('Count')
plt.legend(title='Is Weekend')
plt.show()
```

```python
figure= plt.figure(figsize=(18, 7))
a=sb.barplot(x=df['Month'], y=df['Revenue'], hue=df["Holiday"], palette='coolwarm', edgecolor='none')
plt.title('Monthly Revenue by Holiday')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title='Holiday')
for i in a.containers:
    a.bar_label(i, label_type='edge', fontsize=10, color='black', padding=3)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.lineplot(x='Month', y='Revenue', data=df, marker='o', color='blue')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

```python
figure= plt.figure(figsize=(10, 6))
a=sb.barplot(x=df['City'], y=df['Revenue'], hue=df["Holiday"], palette='coolwarm', edgecolor='none')
plt.title('per city Revenue by Holiday')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.legend(title='Holiday')
for i in a.containers:
    a.bar_label(i, label_type='edge', fontsize=10, color='black', padding=3)
plt.xticks(rotation=45)
plt.show()
```

```python
figure= plt.figure(figsize=(10, 6))
a=sb.barplot(x=df['City'], y=df['Revenue'], palette='coolwarm', edgecolor='none')
plt.title('per city Rtotal Revenue')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.legend(title='Holiday')
for i in a.containers:
    a.bar_label(i, label_type='edge', fontsize=10, color='black', padding=3)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.pie(df['Is_Weekend'].value_counts(), labels=['Weekday', 'Weekend'], autopct='%1.1f%%')
plt.title('Distribution of Weekend vs Weekday Sales')   
plt.show()
```

```python
sb.barplot(x='Store_ID', y='Units_Sold', data=df)
plt.title("Units Sold per Store")
plt.xlabel("Revenue")   
plt.ylabel("Units Sold")
plt.show()
```

```python
sb.boxplot(x='Store_ID', y='Units_Sold', data=df, palette='coolwarm')
plt.title("Box Plot of Units Sold per Store")
plt.xlabel("Store ID")
plt.ylabel("Units Sold")
plt.show()
```

```python
sb.rugplot(x='Revenue', data=df, height=0.1, color='blue')
plt.title('Rug Plot of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Density')
plt.show()
```

```python
drop=df.drop(columns=['Date', 'Store_ID', 'Is_Weekend', 'Holiday','Product_ID','City','Month'])
sb.heatmap(drop.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sb.clustermap(drop.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Clustermap of Correlation')
plt.show()
```
