```python
x=np.linspace(0, 10, 100)
y=x**2
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Plot")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y,"r-.")
plt.subplot(1, 2, 2)
plt.plot(y, x, "b*-")
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, "r")
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")
axes.set_title("Axes Plot")
plt.show(3)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2=fig.add_axes([0.2, 0.5, 0.4, 0.3])
ax1.plot(x, y)
ax1.set_title("Outer Axes")
ax2.plot(y,x)
ax2.set_title("Inner Axes")
ax2.set_xlabel("Inner X-axis")
ax2.set_ylabel("Inner Y-axis")
plt.show(4)
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
plt.show()
```

```python
fig,axis=plt.subplots(nrows=1,ncols=2)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for curr_axes in axes:
    curr_axes.plot(x,y)
plt.show(5) 
```

```python
fig, axes_1 = plt.subplots(nrows=3,ncols=3)

for i in axes_1:
    for j in i:
        j.plot(x,y)
        
plt.tight_layout()
plt.show(6)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.show(7) 
```

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show(8)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4), dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(5,5))
axes.plot(x,x**2)
axes.plot(x,x**3)
plt.show()

axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', lw=3.5, alpha=0.3)# lw is line width, alpha is transparency(0 to 1)
plt.show()
```

```python
plt.scatter(x, y, color='green', marker='o', s=1000) # s is size of the marker  
plt.show()   
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

fig, axes = plt.subplots(1,2, figsize=(12,4))

# Rectangular box plot
axes[0].boxplot(data, vert=True, patch_artist=True)
axes[1].boxplot(data)
plt.show()
```
